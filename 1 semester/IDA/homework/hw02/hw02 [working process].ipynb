# Task A

````python
def ackermann(m, n):
    ackermann.counter = 0
    
    def rec(m, n):
        ackermann.counter += 1
        if m == 0:
            return n + 1
        if n == 0:
            return rec(m - 1, 1)
        return rec(m - 1, rec(m, n - 1))
    
    return rec(m, n)
````

````python
print(ackermann(2, 27), ackermann.counter)
print(ackermann(4, 0), ackermann.counter)
````

````python
import time
import functools


def stopwatch(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        print('`{}` started'.format(func.__name__))
        ret = func(*args, **kwargs)
        print('`{}` finished in {:.2f}s'.format(func.__name__, time.time() - start))
        return ret
    return wrapper
````

````python
import time

@stopwatch
def func123():
    time.sleep(1.25)

if __name__ == '__main__':
    func123()
````

````python
import functools


def once(func):
    called = False
    res = None

    @functools.wraps(func)
    def wrapper(*args, **kwards):
        nonlocal called, res
        if not called:
            res = func(*args, **kwards)
            called = True
        return res
    return wrapper
````

````python
@once
def init():
    print("Settings are successfully set.")
    return 1

if __name__ == "__main__":
    print(init())
    print(init())
````

````python
@once
def bonjour():
    print("Bonjour le monde!")

if __name__ == "__main__":
    bonjour()
    bonjour()
    bonjour()
````

# Task F

````python
def binomial(n):
    for k in range(-1, n):
        Cnk = Cnk * (n - k) // (k + 1) if k >= 0 else 1
        yield Cnk

if __name__ == '__main__':
    for n in map(int, input().split(' ')):
        print(' '.join([str(x) for x in binomial(n)]))
````

# Task G

````python
def check(s):
    delta = 0
    for c in s:
        delta += 1 if c == '(' else -1
        if delta < 0:
            return False
    return delta == 0
    
def brackets(n):
    def rec(n, s):
        if n == 0:
            if check(s):
                yield s
            return
        yield from rec(n - 1, s + '(')
        yield from rec(n - 1, s + ')')
    yield from rec(2 * n, '')

if __name__ == '__main__':
    print('\n'.join([str(x) for x in brackets(int(input()))]))
````
